```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.ollama import Ollama
```

I have put the pdf with the full application in the directory data.

```python
documents = SimpleDirectoryReader("data").load_data()
```

```python
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
llm = Ollama(model="mistral", request_timeout=360.0)
```

```python
index = VectorStoreIndex.from_documents(documents, embed_model=embed_model)
query_engine = index.as_query_engine(llm=llm)
```

```python
response = query_engine.query("Wie lautet das Ohmsche Gesetz? Benutze lediglich das gelesene Dokument. Gib die Text Passagen an, aus denen du die Antwort ableitest.")
print(response)
```
