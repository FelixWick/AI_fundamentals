```python
import ollama
from typing import List, Dict
```

```python
def call_llm(prompt: str) -> str:
    response = ollama.chat(model="mistral", messages=[
        {"role": "user", "content": prompt}
    ])
    return response["message"]["content"].strip()
```

```python
def generate_plan(goal: str) -> List[str]:
    prompt = f"""
You are a planning agent. Break down the following goal into numbered small steps:

Goal: {goal}
"""
    plan_text = call_llm(prompt)
    
    # Convert numbered list into Python list
    steps = []
    for line in plan_text.split("\n"):
        line = line.strip()
        if line[:1].isdigit():
            step = line.split(" ", 1)[1]
            steps.append(step)
    return steps
```

```python
def execute_step(step: str, state: Dict) -> str:
    prompt = f"""
You are an execution agent. Given the current state:

State: {state}

Execute this step:
- {step}

Return the result only.
"""
    return call_llm(prompt)
```

```python
def run_agent(goal: str):
    print(f"=== Goal: {goal} ===\n")

    plan = generate_plan(goal)
    print("Generated Plan:")
    for i, step in enumerate(plan, 1):
        print(f"  {i}. {step}")

    state = {}
    print("\nExecuting...\n")

    for i, step in enumerate(plan, 1):
        result = execute_step(step, state)
        state[f"step_{i}"] = result
        print(f"[Step {i}] {step}")
        print(f"Result: {result}\n")

    print("=== Final State ===")
    print(state)
    return state
```

```python
run_agent("Plan a 3-day trip to Tokyo with food and sightseeing recommendations.")
```
