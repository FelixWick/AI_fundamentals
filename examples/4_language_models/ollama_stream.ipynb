```python
import requests
import json
```

```python
def stream_ollama(model: str, prompt: str):
    """Stream Ollama model output in real time."""
    url = "http://localhost:11434/api/generate"
    payload = {"model": model, "prompt": prompt, "stream": True}

    # stream=True allows reading as the model generates
    with requests.post(url, json=payload, stream=True) as response:
        response.raise_for_status()
        print("\n--- Response ---\n")
        for line in response.iter_lines():
            if not line:
                continue
            data = json.loads(line.decode("utf-8"))
            if "response" in data:
                print(data["response"], end="", flush=True)
            if data.get("done"):
                break
        print("\n\n--- Stream complete ---")
```

```python
model = "llama3"
prompt = "Describe how neural networks learn, in one paragraph."
stream_ollama(model, prompt)
```
