```python
from ultralytics import YOLO
```

```python
detection_model = YOLO("yolov8n.pt")
detection_model.predict("https://ultralytics.com/images/bus.jpg", conf=0.3, show_boxes=True, show_labels=True, show_conf=True, save=True, save_dir="yolo/runs/detect")
```

```python
segmentation_model = YOLO("yolov8n-seg.pt")
segmentation_model.predict("https://ultralytics.com/images/bus.jpg", conf=0.2, show_boxes=True, show_labels=True, show_conf=True, save=True, save_dir="yolo/runs/segment")
```
