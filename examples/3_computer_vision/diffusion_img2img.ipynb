```python
import torch
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
```

```python
# Choose a model from Hugging Face
model_id = "runwayml/stable-diffusion-v1-5"  # You can change this to another model
```

```python
device = (
    torch.device("cuda") if torch.cuda.is_available()
    else torch.device("mps") if torch.backends.mps.is_available()
    else torch.device("cpu")
)
print(f"Using device: {device}")
```

```python
# Load the pipeline
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(device)
```

```python
# Load your input image
init_image = Image.open("generated_image.png").convert("RGB")
```

```python
# Set your prompt
prompt = "Turn this into a fantasy landscape painting"
```

```python
# Generate the image
# strength: how much to transform the input (0 = keep original, 1 = fully new)
image = pipe(
    prompt=prompt,
    image=init_image,
    strength=0.7,           # 0.0â€“1.0
    guidance_scale=7.5,     # higher = more faithful to prompt
    num_inference_steps=50
).images[0]
```

```python
# Save the generated image
image.save("img2img_result.png")
print("Image saved as img2img_result.png")
```
