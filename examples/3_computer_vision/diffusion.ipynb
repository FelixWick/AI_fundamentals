```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
# Choose a model from Hugging Face
model_id = "runwayml/stable-diffusion-v1-5"  # You can change this to another model
```

```python
device = (
    torch.device("cuda") if torch.cuda.is_available()
    else torch.device("mps") if torch.backends.mps.is_available()
    else torch.device("cpu")
)
print(f"Using device: {device}")
```

```python
# Load the pipeline
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(device)
```

```python
# Generate an image from a prompt
prompt = "A scenic view of mountains at sunset, digital art"
num_inference_steps = 50  # Higher = better quality, slower
guidance_scale = 7.5      # Higher = more faithful to prompt

image = pipe(prompt, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale).images[0]
```

```python
# Save the generated image
image.save("generated_image.png")
print("Image saved as generated_image.png")
```
